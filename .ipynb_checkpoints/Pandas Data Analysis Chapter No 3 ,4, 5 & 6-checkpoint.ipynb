## Chapter 5 - Apply

```python
students_details = pd.read_csv('student-mat.csv')
students_details
```

```python
students_details['Medu'] = students_details['Medu'] + 2
students_details.head(10)
```

```python
students_details['Mjob'] = students_details['Mjob'].str.upper()
students_details.tail(10)
```

```python
students_details['guardian'] = students_details['guardian'].apply(func = lambda x:x.capitalize())
students_details
```

```python
students_details['guardian']
```

### Create a new column 'Eligibility_Criteria' where it returns 1 if Age>=18 else 0.

### Apply function using lambda function

```python
students_details['Eligibility_Criteria'] = students_details['age'].apply(lambda age:1 if age>=18 else 0)
```

```python
students_details
```

```python
del students_details['Eligibility_Criteria']
```

```python
students_details
```

### Apply function using user defined function

```python
def get_age(x):
    if x>=18:
        return 1
    else:
        return 0
```

```python
students_details['Eligibility Criteria'] = students_details['age'].apply(get_age)
students_details
```

## Chapter 6 - Concat | Merge | Append

```python
sales_Transactions_2017 = pd.read_csv('Sales Transactions-2017.csv')
sales_Transactions_2017
```

```python
sales_Transactions_2018 = pd.read_csv('Sales Transactions-2018.csv')
sales_Transactions_2018
```

```python
sales_Transactions_2019 = pd.read_csv('Sales Transactions-2019.csv')
sales_Transactions_2019
```

```python
sales_complete_data = pd.concat([sales_Transactions_2017,sales_Transactions_2018,sales_Transactions_2019])
sales_complete_data
```

```python
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
## or we can use append for same purpose
sales_Transactions_2017.append([sales_Transactions_2018,sales_Transactions_2019])
```
